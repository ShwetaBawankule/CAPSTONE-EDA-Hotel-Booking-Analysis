```python
#Creating a copy of orignial data for renaming few column
hotel_df=hotel_booking_df.copy()
```

```python
#Renaming the 'adr' column name to 'Average_daily_rate' for understanding the data well.
hotel_df.rename(columns={'adr':'Average_daily_rate'},inplace=True)
```

```python
# Numerical information of dataset
hotel_df.describe()
```

```python
# Total number of null values in each column
hotel_df.isnull().sum()
```

```python
# 'agent'&'company' row has huge number of null values and these are also not required for our analysis so we will remove it
new_hotel_df=hotel_df.drop(['agent','company'],axis=1)
new_hotel_df
# new_hotel_df is the dataframe which doesn't include agent & company columns
```

```python
#Let's work on two columns country & children we are'nt skipping this two columns ,
# as we did for agent & company cause the missing data is small as compared to it .
```

```python
new_hotel_df['country'].fillna(new_hotel_df['country'].mode().to_string(),inplace=True) #Replace the null values of country with mode value of the column
new_hotel_df['children'].fillna(round(new_hotel_df['children'].mean()),inplace=True) #Replace the null values of children with mean value of the column
```

```python
# Deleting the rows with empty cells
new_hotel_df=new_hotel_df.dropna(axis=0)
```

```python
# Re-checking the number of null values to ensure the data cleaning is successful.
new_hotel_df.isnull().sum()
```
