```python
#Creating a copy of orignial data for renaming few column
hotel_df=hotel_booking_df.copy()
```

```python
#Renaming the 'adr' column name to 'Average_daily_rate' for understanding the data well.
hotel_df.rename(columns={'adr':'Average_daily_rate'},inplace=True)
```

```python
# Numerical information of dataset
hotel_df.describe()
```

```python
# Total number of null values in each column
hotel_df.isnull().sum()
```

```python
# 'agent'&'company' row has huge number of null values and these are also not required for our analysis so we will remove it
new_hotel_df=hotel_df.drop(['agent','company'],axis=1)
new_hotel_df
# new_hotel_df is the dataframe which doesn't include agent & company columns
```

```python
#Let's work on two columns country & children we are'nt skipping this two columns ,
# as we did for agent & company cause the missing data is small as compared to it .
```

```python
new_hotel_df['country'].fillna(new_hotel_df['country'].mode().to_string(),inplace=True) #Replace the null values of country with mode value of the column
new_hotel_df['children'].fillna(round(new_hotel_df['children'].mean()),inplace=True) #Replace the null values of children with mean value of the column
```

```python
# Deleting the rows with empty cells
new_hotel_df=new_hotel_df.dropna(axis=0)
```

```python
# Re-checking the number of null values to ensure the data cleaning is successful.
new_hotel_df.isnull().sum()
```

```python
 # We observe that there are no more null values in the dataset so, the data is cleaned.
```

# We observe that there are no more null values in the dataset so, the data is cleaned.

```python
#Creating a column with totalnightstays
new_hotel_df['total_night_stays']=new_hotel_df['stays_in_weekend_nights']+new_hotel_df['stays_in_week_nights']
new_hotel_df
```

## 4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables.

Chart 1-Types of hotel in the dataset and most preffered among those by the customers

```python
most_preffered_hotel = new_hotel_df['hotel'].value_counts()
most_preffered_hotel  # Two types of hotels are there in dataset :'City Hotel'&'Resort Hotel'
```

```python
#Percentage of hotel preffered
new_hotel_df['hotel'].value_counts(normalize=True)*100
```

```python
# Let's Visualize the above data
most_preffered_hotel.plot(kind='pie',autopct='%1.0f%%')
plt.title('Type of Preferred Hotel')
plt.show
```

```python
# I have choosen pie-chart to show how much percentage does each hotel type contribute to the whole
# The pie chart reveals that 66.4% of the hotels in the dataset are 'City Hotels' and the other 33.6% are 'Resort Hotels'.
```

Chart 2: From which countries most of the customers visit these hotels

```python
grp_by_country=new_hotel_df.groupby('country')
d2 = pd.DataFrame(grp_by_country.size()).rename(columns = {0:'no. of bookings'}).sort_values('no. of bookings', ascending = False)
d2 = d2[:10]
sns.barplot(x = d2.index, y = d2['no. of bookings'])
plt.show()
```

Most of the customers come from Portugal, Great Britain, France and Spain.

Chart3: Which was the most preferred year for hotel booking

```python
#Code for most preffered year by the customer for hotel booking.
most_preffered_year=new_hotel_df['arrival_date_year'].value_counts()
most_preffered_year

# We can see number of bookings which took place in the following years 2015,2016,2017
```

```python
# Let's visualize the data for the most preffered year
most_preffered_year.plot(kind='bar')
plt.title('Year of preference')
plt.xlabel('Year')
plt.ylabel('Number of bookings')
plt.show()
```

```python
# We can see that 2016 was the most preferred year by the customers,followed by the 2017 & 2015
```

Chart 4 :Which was the most preferred month by the customers for the bookings


```python
#Let's code for the most_preferred_month
most_preferred_month=new_hotel_df['arrival_date_month'].value_counts()
most_preferred_month
```

```python
#Let's visualize for the most preferred month
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
most_preferred_month.index = pd.Categorical(most_preferred_month.index, categories=months, ordered=True)
most_preferred_month.sort_index(ascending=True).plot(kind='bar')
plt.title('Most preferred month')
plt.xlabel('Months')
plt.ylabel('No of bookings')
plt.show()
```

```python
# From this graph we can conclude that August month has highest no of bookings & January has the lowest  number of bookings
```

Chart 5 :What is the monthly booking of the hotels?

```python
# Let's code for the type of hotel and its booking count in the preferred months
monthly_booking_hotels=new_hotel_df.groupby(['hotel','arrival_date_month'])['hotel'].count().unstack()
monthly_booking_hotels
```

```python
# Let's visualize it with graphs
monthly_booking_hotels.plot(kind='bar',fontsize =12,figsize=(15,7))
plt.title('Monthly Hotel Bookings')
plt.xlabel('Months')
plt.ylabel('No of bookings')
plt.show()
```

Chart 6 :Preferrence of hotel with respect to year

```python
# Yearly Booking of hotels
yearly_booked_hotels = new_hotel_df.groupby(['arrival_date_year','hotel'])['hotel'].count().unstack()
yearly_booked_hotels
```

```python
#Let's visualize the yearly booked hotels analysis
yearly_booked_hotels.plot(kind='bar')
plt.title('Yearly Booked Hotels')
plt.xlabel('Year')
plt.ylabel('Number of bookings')
plt.show()
```

Chart 7:Anaylsis on cancellation of hotels


```python
# Cancellation data of hotels(0 = Not cancelled & 1 = Cancelled)
cancellation_data=new_hotel_df.groupby(['is_canceled','hotel'])['is_canceled'].count().unstack()
cancellation_data
```

```python
#Let's visualize the anaylsis on hotel cancellation
cancellation_data.plot(kind='bar')
plt.title('''Cancellation status

0 = Not cancelled & 1 = Cancelled''')
plt.xlabel('Hotel')
plt.ylabel('Count')
plt.show()
```

Chart 8:Cancellation in percentage pie-chart for City Hotel

```python
# Cancellation percentage of 'City Hotel'.
city_hotel_cancellation = new_hotel_df[new_hotel_df['hotel'] == 'City Hotel']
cancellation_city_hotel = city_hotel_cancellation['is_canceled'].value_counts(normalize=True)*100
cancellation_city_hotel
```

```python
# Plotting it in pie chart.
cancellation_city_hotel.plot(kind='pie',autopct ='%1.1f%%',figsize =(8,8),fontsize = 15,wedgeprops = {"edgecolor" : "white",
                      'linewidth': 10,
                      'antialiased': True})
plt.title('''CITY HOTEL CANCELLATION STATUS

(0 = Not cancelled & 1 = Cancelled)''')
plt.show()
```

Chart 9:Cancellation in percentage pie-chart for Resort Hotel

```python
# Cancellation percentage of 'Resort Hotel'.
resort_hotel_cancellation = new_hotel_df[new_hotel_df['hotel'] == 'Resort Hotel']
cancellation_resort_hotel = resort_hotel_cancellation['is_canceled'].value_counts(normalize=True)*100
cancellation_resort_hotel
```

```python
# Plotting it in pie chart.
cancellation_resort_hotel.plot(kind='pie',autopct ='%1.1f%%',figsize =(8,8),fontsize = 15)

plt.title('''RESORT HOTEL CANCELLATION STATUS

(0 = Not cancelled & 1 = Cancelled)''')
plt.show()
```

Chart 12:Frequency of repeated guest
